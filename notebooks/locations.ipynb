```python
def fillRegion(row):
    if pd.isnull(row["region"]):
        match row["city"]:
            case "GUAYAQUIL":
                return "GUAYAQUIL"
            case "QUITO":
                return "QUITO"
            case "MANTA":
                return "MANABI"
            case _:
                return "OTROS"
    else:
        return row["region"]
```

```python
def fillCommerce(row):
    if pd.isnull(row["commerce"]):
        match row["type"]:
            case "BODEGA":
                return "BODEGA"
            case _:
                return "OTROS"
    else:
        return row["commerce"]
```

```python
def SetBuildingSize(row):
    if row["size"] < Q2:
        return "PEQUEÑO"
    elif row["size"] < Q3:
        return "MEDIANO"
    else:
        return "GRANDE"
        
```

```python
df_loc.drop(columns=["latitude","longitude"],inplace=True)
df_loc.head()
```

```python
df_loc["region"]=df_loc.apply(fillRegion, axis=1)
```

```python
df_loc.loc[(df_loc['name'].str.startswith("BOD")) & (df_loc['type'].isna()), 'type'] = "BODEGA"
df_loc.loc[(df_loc['type'].isna()), 'type'] = "OTROS"
```

```python
df_loc["commerce"]=df_loc.apply(fillCommerce, axis=1)
df_loc.isnull().sum()
```

```python
#Verifico los que quedan y compruebo que son de la ciudad de Guayaquil
df_loc[(df_loc["city"].isnull())]
```

```python
df_loc.loc[(df_loc["city"].isnull()),"city"]="GUAYAQUIL"
df_loc.loc[(df_loc["state"].isnull()) & (df_loc['region']=="OTROS"),"region"]="GUAYAQUIL"
df_loc.loc[(df_loc["state"].isnull()),"state"]="GUAYAS"
df_loc.isnull().sum()
```

```python
df_loc["tamanio"]=df_loc.apply(SetBuildingSize, axis=1)
```

```python
df_loc.to_csv("../output/locations_cleaned.csv", index=False)
```
